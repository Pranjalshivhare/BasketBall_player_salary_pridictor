```python
#for salary prediction, full_name, birthday, height, weight, college 
#are not the important factors to be considered.
#droping these features.
df = df.drop(['full_name','b_day','height','weight','college','version'], axis=1)
```

```python
#we don't need '#' in jersey column
#creating a function to remove '#'

def removeHash(value):
    value = value[1:]
    return int(value)

df['jersey'] = df['jersey'].apply(removeHash)
```

```python
df.head()
```

```python
#we don't need '$' in salary column
#creating a function to remove '$'

def removeDollor(value):
    value = value[1:]
    return int(value)

df['salary'] = df['salary'].apply(removeHash)
```

```python
#for working with data,we need to convert string values to labels.
#example, TeamName, country and postion are present in string format.
#We need to convert these into labels.
```

```python
#cheking how many unique values are there in the team column
df['team'].value_counts()
```

```python
df['country'].value_counts()
```

```python
#from the country counts, we can figure out that except
#USA                   315
#Canada                 15
#Australia               8
#Cameroon                5
#Nigeria                 3
#The Bahamas             2
#Ukraine                 2

#all other country have 1 - 1 player each
# we can wrap all these players with the country name "others"

def removeCountry(value):
    if value not in ['USA','Canada','Australia','Cameroon','Nigeria','The Bahamas','Ukraine']:
        return 'Others'
    else:
        return value
    
df['country'] = df['country'].apply(removeCountry)
df['country'].value_counts()
```

```python
df['position'].value_counts()
```

```python
df['team'].value_counts()
```

```python
df['draft_round'].value_counts()
df['draft_peak'].value_counts()
```

```python
#From above data, draft_peak and draft_round contain undrafted data
# removing undrafted data, and rows from final data frame

def removeUndrafted(value):
    if value=='Undrafted':
        return
    else:
        return value
df['draft_round'] = df['draft_round'].apply(removeUndrafted)
df['draft_peak'] = df['draft_peak'].apply(removeUndrafted)

df = df.dropna()
```

```python
df['draft_peak'].value_counts()
```

```python
#labeling of column name country, team and position.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['position'] = le.fit_transform(df['position'])
df['country'] = le.fit_transform(df['country'])
df['team'] = le.fit_transform(df['team'])

df.head()
```

```python
x = df.drop(['salary'], axis=1)
y = df['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```
